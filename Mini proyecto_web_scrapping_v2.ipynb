```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import seaborn as sns
'''
APUNTES
Ejemplo:
req=requests.get('http://www.xxxxxxx.com')
req.status_code
req.headers <head> </haed>   ->METADATA
req.text  <body> </body>     ->CONTENIDO
'''
```

```python
#Pagina web: https://www.pdpaola.com/es

url='https://www.pdpaola.com/es/collections/rings'
headers={'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

response = requests.get(url,headers=headers)

response.status_code
```

```python
response.text
```

```python
type(response.text)
```

```python
soup = BeautifulSoup(response.text)
soup
```

```python
type(soup)
```

```python
lista_anillos=soup.find_all(name="div",class_="product-item-title")
lista_anillos
```

```python
def extrae_nombre(list_names):
    '''
    Esta función extrae el nombre del anillo
    '''
    lista_nombres_final=[]
    for i in list_names:
        lista_nombres_final.append(i.getText().replace('\n',''))
    return lista_nombres_final
```

```python
extrae_nombre(lista_anillos)
```

```python
lista_descripcion=soup.find_all("div",{'class':"product-item-subtitle"})
lista_descripcion
```

```python
def extrae_descripcion(list_description):
    '''
    Esta función extrae la descripción del anillo
    '''
    lista_descripcion_final=[]
    for i in list_description:
        lista_descripcion_final.append(i.getText().replace('\n','   ').strip().replace('   ',' / ')) #.replace('\n','')
    return lista_descripcion_final
```

```python
extrae_descripcion(lista_descripcion)
```

```python
lista_precios=soup.find_all("div",{'class':"product-price product-item-price"})
lista_precios
```

```python
def extrae_precio(list_price):
    '''
    Esta función extrae el precio del anillo
    '''
    lista_precio_final=[]
    for i in list_price:
        lista_precio_final.append(i.getText().replace('\n',' ').replace('€','').replace(',00','').strip().split(' ')[0])
    return lista_precio_final
```

```python
extrae_precio(lista_precios)
```

```python
ring_dicc=dict(Nombre_anillo=extrae_nombre(lista_anillos),
               Descripcion=extrae_descripcion(lista_descripcion),
               Precio=extrae_precio(lista_precios)
              )
ring_dicc
```

```python
ring_df=pd.DataFrame(ring_dicc)
ring_df
```

```python
# Extraer csv
ring_df.to_csv('Ring_data.csv')
```
